```python
processed_date = ''
batch_month = ""
```

# **reading customer dataframe, store in a variable, insert a new column (processing_date), then write to bronze layer**

```python
from pyspark.sql.functions import lit

cust_df = spark.read.format("csv").option("header","true").load(f"Files/landing/customers-{batch_month}.csv")

row_count = cust_df.count()

if row_count > 0:
    print(f"the file contains data - {row_count} rows found")
    
    # add processing_date for partitioning purposes
    customers_df = cust_df.withColumn("processing_date", lit(processed_date))

    # writing to bronze as delta table
    customers_df.write.format('delta').partitionBy("processing_date").mode("append").save("Files/bronze/customers")
    print(f"data successfully written to bronze : {row_count} rows inserted")

else:
    print("the file contains no data")
```
